```python
from email.mime import image

# open('파일경로', 모드(r,w,a), **options)
# - 지정된 경로에 파일을 여는 파이썬 내장 함수
# - 열린 파일에 지정된 모드 동작 수행을 할 수 있다.

f = open('helloworld.txt','r',encoding='utf-8')
#'r' : read 모드
# encoding='utf-8' : 한글도 읽을 수 있게
text = f.read()
print(text)
f.close()
```

## 파일 쓰기

```python
f=open('helloworld2.txt','w',encoding='utf-8')
#'w' : write mode
print(f,type(f))

text = """
아
내일이
화요일이라니...
hahaha...
"""

f.write(text)
#지정된 파일에 text 출력
# 기존 내용이 있으면 덮어쓰기
# 지정된 파일이 존재하지 않으면 파일 생성

# 파일 연결에 사용된 메모리 반환
f.close()
```

```python
# with절
# - with 옆에 작성된 open() 구문에서 사용한 파일 연결을 with절 종료 시 자동으로 close()

with open('helloworld2.txt','w',encoding='utf-8') as f:
    f.write("with절을 이용하면 자동 close() 된다.")
```

## 파일 읽기

```python
with open('helloworld.txt','r',encoding='utf-8') as f:
    print(f.read())     #전부 읽어와서 출력
```

```python
# 한 줄씩 읽기
with open('helloworld.txt','r',encoding='utf-8') as f:
    for i, line in enumerate(f):
        print(i+1, line, end='')
        # end='' : print() 완료 후 마지막 문자 빈칸 처리 (엔터가 기본값)
        # i+1 : 인덱스 숫자 +1 (기본값은 인덱스 0부터 시작)
```

```python
# 파일 이어쓰기 ('a', append 모드)
with open('helloworld.txt','a',encoding='utf-8') as f:
    f.write(''' ㅋㅋㅋ
🤢🤮🤕
☠️☠️☠️
👹
💩''')

with open('helloworld.txt','r',encoding='utf-8') as f:
    print(f.read())     #이어쓰기 확인
```

```python
# 파일 복사
# 1. 파일 읽기
with open('helloworld.txt','r',encoding='utf-8') as f:
    text=f.read()       #with절에 선언한 변수는 지역변수X -> 전역 사용 가능

# 2. 파일 쓰기
with open('helloworld_copy.txt','w',encoding='utf-8') as f:
    f.write(text)
    print("파일 복사 완료")
```

```python
# 이미지 파일 복사
# 파일 복사는 아무 파일이나 가능
with open('똥꼬푸린.jpeg','rb') as f:
    # 'r', 'w', 'a' : 문자 전용 읽기, 쓰기, 이어쓰기 모드
    # 'rb', 'wb' : 문자가 아닌 파일은 byte 단위로 읽기, 쓰기 해야함
    #print(f.read())
    image_bytes = f.read()

with open('copy_image.jpeg','wb') as f:
    f.write(image_bytes)
```

## dict-csv 읽고 쓰기

```python
# csv 읽기
# csv.DictReader(f)
# csv.DictWriter(f, fieldnames)
import csv

with open('users.csv','r',encoding='utf-8') as f:
    reader=csv.DictReader(f)    #DictReader 객체 생성

    users=[]

    # for row in reader:
    #     print(row,type(row))
    #     users.append(row)

    users = [row for row in reader]

    print(users)
```

```python
# csv 쓰기
with open('users.csv','w',encoding='utf-8',newline='') as f:
    # csv 쓰기 객체 생성
    # - f : 작성할 파일
    # - fieldnames : 첫 줄에 작성될 헤더(칼럼명)
    writer=csv.DictWriter(f,fieldnames=['id','name','email'])
    writer.writeheader()        # fieldnames로 헤더 작성

    # users에 사용자 추가
    users.append({'id':400, 'name':'이순신', 'email':'sslee@gmail.com'})
    users.append({'id':500, 'name':'강감찬', 'email':'gck@gmail.com'})

    # 여러 줄 쓰기
    # writer.writerows(users)

    # 한 줄씩 쓰기
    for user in users:
        writer.writerow(user)
```
