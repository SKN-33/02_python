raise 예외객체를 사용해서 명시적으로 예외를 던질수 있다.
이를 통해 코드 처리흐름을 명확히 전달할 수 있다.

```python
def foo():
    raise NotImplementedError("foo() 함수가 구현되지 않았습니다.")

foo()
```

```python
class AdultGame:
    def __init__(self):
        pass
    def start(self):
        print("성인용 게임(고스톱)을 시작합니다.")

# if - else
age = int(input("나이 입력: "))
if age >= 19:
    AdultGame().start()
else:
    print("돌아가")
# if - else가 잘못된 것은 아니지만
# 코드 흐름 의도와 맞지는 않다
```

```python
# 사용자 정의 예외 클래스
# - 만드는 방법: 기존 Exception, Error 클래스 중 아무거나 상속 받기
class UnderAgeError(Exception):
    pass

class AdultGame2:
    def __init__(self, age: int):
        if age < 19:
            # BaseException : 모든 예외, 에러 클래스의 최상위 부모 클래스
            # raise BaseException(f'미성년자는 이용할 수 없습니다: {age}') = 권장X

            # 상황에 맞는 예외 강제 발생
            raise UnderAgeError(f'미성년자는 이용할 수 없습니다: {age}')

    def start(self):
        print("성인용 게임(고스톱)을 시작합니다.")

try:
    age = int(input("나이 입력: "))
    AdultGame2(age).start()
except UnderAgeError as e:
    print(e)

print("종료")
```
